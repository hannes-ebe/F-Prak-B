Besetzungswahrscheinlichkeiten der Schwingungsniveaus im elektronischen Grundzustand

```python
import numpy as np
import scipy.constants as c
import matplotlib.pyplot as plt
```

```python
v_0 = 214.6*1e2
v_0x = 0.613*1e2
T = 293.15
```

```python
def kanonische_Zustandssumme(E,T,index_max):
    '''E als Liste/array mit den entsprechenden Energien'''
    sum = 0.0
    for v in range(index_max):
        sum += np.exp(-E[v]/c.k/T)
    return sum
def besetzungswahrscheinlichkeit(Z,v,T):
    return np.exp(-Energieeigenwerte(v)/c.k/T)/Z
def Energieeigenwerte(v):
    return c.h*c.c*v_0*(v+0.5)-c.h*c.c*v_0x*(v+0.5)**2
```

```python
v = [Energieeigenwerte(i) for i in range(250)]
v_max = max(v)
index_max = v.index(v_max)
print(v_max)
print(index_max)
print(v[index_max])
print(Energieeigenwerte(index_max))
```

```python
E = [Energieeigenwerte(v) for v in range(index_max+1)]
Z = kanonische_Zustandssumme(E,T,index_max)
p_0 = besetzungswahrscheinlichkeit(Z,0,T)
p_1 = besetzungswahrscheinlichkeit(Z,1,T)
p_2 = besetzungswahrscheinlichkeit(Z,2,T)
print('Besetzungswahrscheinlichkeit v=0: ', p_0)
print('Besetzungswahrscheinlichkeit v=1: ', p_1)
print('Besetzungswahrscheinlichkeit v=2: ', p_2)
```
