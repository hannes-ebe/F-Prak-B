```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
class spek:
    def __init__(self,path):
        data=np.loadtxt(path)
        self.steps=[]
        self.intensity=[]
        for line in data:
            self.steps.append(line[0])
            self.intensity.append(line[1])

    def plot(self):
        fig = plt.figure(figsize=(11, 6))
        gs = GridSpec(8, 5)
        fig1 = fig.add_subplot(gs[:, :])
        fig1.set_title("Spektrum")
        fig1.set_ylabel("step")
        fig1.set_xlabel("Intensity")
        fig1.plot(self.steps,self.intensity,label=f"plot")
        plt.tight_layout()
        plt.legend()
        plt.show()
```

```python
file="1_-4_ohneLinse.dat"
spek1=spek(f"data/{file}")
spek1.plot()
```
